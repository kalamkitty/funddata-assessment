# Normal Distribution

This notebook is about the normal distribution, it is the most common distribution in statistics as it is used to represent random variables with unknown distributions. 


## Definition
* * *
The normal distribution is a "continuous probability distribution that is symmetrical around its mean, with most of observations clubster around central peak" and other probabilites of variables taper on both sides of the distribution (Frost, 2018). It is used to describe how variables are spread/distributed for many natural phenomenas for example IQ, height, income distribution and student average test scores (Malik, 2019).

The function of a normal distribution is:



*f(x)* = probability density function

$\sigma$= standard deviation

$\mu$ = mean

## Characteristics of Normal Distribution
* * *


Although normal distributions can have different shapes, there are a many characteristics that are common amongst normal distributions.

- *Mean, median and mode are equal*, the mean/center of the normal distribution is the value of maximum frequency.

- All curves are *symmetric* around the mean in a bell shape, half of the values are below the meanand the other half above the mean, therefore   the skewness is zero. 

- *Empirical rule*, the values of a normal distribution are always a specific number of standard deviations from the mean. This is explained further below.

### Mean, Median and Mode

Using python we can calculate the mean, median and mode.
Mean is the average value, median is the mid point value and mode is the most common value.

```python
# https://www.w3schools.com/python/python_ml_mean_median_mode.asp
import numpy as np
import scipy.stats as stats

# Generate 100 random data point between 1-10
data= np.random.randint (1,10,100).round()
print (data)
mean = np.mean (data)
print ("Mean of data is :" f"{mean}")
```

```python
median= np.median (data)
print ("Median of data is :" f"{mean}")
```

```python
mode= stats.mode (data)
print ("Mode of data is :" f'{mode[0]}')
```

### Empirical Rule

The empirical rule is also know as the 68 95 99 rule. It states that "68% of observed data points will lie inside one standard deviation of the mean, 95% will fall within two standard deviations, and 99.7% will occur within three standard deviations"

From the normal distribution above, 68% is represented in pink, 95% is represented in pink and purple and 99.7% is represented by pink, purple and green.

It is easy to identity any outliers in values in a data set as 99.7% of all observations should be within three standard deviations of the mean.

To test the empirical rule, we can use the cdf() function from scipystats.norm module to calculate the area under a normal distribution curve. The results are roughly the percentages given above.

```python
import matplotlib.pyplot as plt
from scipy.stats import norm
 
# Setting mean as 0 and standard deviation as 1
mean = 0
SD = 1
 
# value of cdf between one, two and three S.D. around the mean
one_sd = norm.cdf (SD, mean, SD) - norm.cdf (-SD, mean, SD)

two_sd = norm.cdf (2 * SD, mean, SD) - norm.cdf (-2 * SD, mean, SD)

three_sd = norm.cdf (3 * SD, mean, SD) - norm.cdf (-3 * SD, mean, SD)
 
# printing the value of fractions within each band
print ("Fraction of values within one SD =", one_sd)
print ("Fraction of values within two SD =", two_sd)
print ("Fraction of values within three SD =", three_sd)
```

## Parameters
* * *
1. Mean
   
   The mean ($\mu$) is the average value of the dataset/variables. It is calculated by the sum of total values divided by the total number of values. It is the center of the normal distrubtion and where the peak is on the bell curve. Using python, I will show normal distributions with different means but with the *same* standard deviation. The changes in mean shifts the curve along the x-axis.
 

```python
# https://vitalflux.com/normal-distribution-explained-python-examples/
from scipy import stats

# Plot multiple normal distribution, generate 1000 random values between -100 and 100.
x = np.linspace(-100, 100, 1000)

# Mean of 1, stnd dev of 1
plt.plot(x, stats.norm(1, 1).pdf(x))

# Mean of 3, stnd dev of 1
plt.plot(x, stats.norm(3, 1).pdf(x))

# Mean of 6, stnd dev of 1
plt.plot(x, stats.norm(6, 1).pdf(x))

plt.xlim(-10, 10)
plt.title('Normal Distributions with different means', fontsize='15')
plt.xlabel('Values of Random Variable X', fontsize='15')
plt.ylabel('Probability', fontsize='15')
plt.legend(["M=1, SD=1","M=3, SD=1","M=6, SD=1"])
plt.show()
```

2. Standard deviation

   The standard deviation ($\sigma$) describes how "spread out" the normal distribution is. The bigger the standard deviation, the flatter the curve. 
   
   The formula to calculate standard deviation from a sample is:
   
   
   or easier using the numpy library in python with the built in std() method.

```python
# https://numpy.org/doc/stable/reference/random/generated/numpy.random.randint.html
# Randomly generating 100 int from -100 to 100, rounding
randnumbers= np.random.randint (-100,100,100).round()

# Find the standard deviation of the randomly generated values
stnddev = np.std(randnumbers)

print(stnddev)
```

   The below code creates plots to show normal distributions with the same mean but different standard deviation. The smaller the standard deviation, the narrower the curve as seen on the blue distribution below.

```python
# Mean of 1, stnd dev of 1
plt.plot(x, stats.norm(1, 1).pdf(x))

# Mean of 1, stnd dev of 2
plt.plot(x, stats.norm(1, 2).pdf(x))

# Mean of 1, stnd dev of 4
plt.plot(x, stats.norm(1, 4).pdf(x))

plt.xlim(-10, 10)
plt.title('Normal Distributions with different standard deviations', fontsize='15')
plt.xlabel('Values of Random Variable X', fontsize='15')
plt.ylabel('Probability', fontsize='15')
plt.legend(["M=1, SD=1","M=1, SD=2","M=1, SD=3"])
plt.show()
```

# Standard Normal Distribution 

Below is a random standard normal distribution created using numpy and matplotlib on python.

```python
# https://numpy.org/doc/stable/reference/random/generated/numpy.random.normal.html
# Random sample with mean of 0, SD of 1 size 100
sample = np.random.normal (0,1,10000)

plt.hist (sample, bins= 30, density= True, linewidth=2)

plt.title("Standard Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```
